# 미스터리 사인
- 두 숫자의 연산을 통해 도출된 결론을 보고 두 숫자가 어떤 연산(계산)을 거쳤는지 추론하기
- 수학적인 계산으로 이루어진 규칙만 해결 가능

## Q1.
- 22 ? 7 = 37
- 큰 수 + 두 수의 차

```python
# 머신러닝이 규칙을 찾게 하기 위해 필요한 것
# 1. 문제 데이터 : 22,7
# 2. 정답 데이터 : 37
# 3. 사용할 머신러닝 모델
# 참고사항 : 1번과 2 번이 많을수록 좋다
```

```python
# 데이터 만들기 = 데이터 수집하기
# 1번 숫자? 2번 숫자 = 정답
# 1번 숫자 : 1~100
# 2번 숫자 : 1~100
import pandas as pd
list1 = []
for i in range(1,101):# 1번 숫자
    for j in range(1,101): # 2번 숫자
        result = (i+(i-j))if i > j else (j+(j-i))
        list1.append([i,j,result])
        
```

```python
# 데이터 전처리
# 머신러닝이 학습 가능한 형태로 데이터 변경
# 배열간의 연산이 가능한 형태 : numpy.array, pandas.Series, pandas.DataFrame
# 같은 위치끼리 수학적인 계산이 가능한 형태
data1 = pd.DataFrame(list1, columns = ['N1', 'N2', 'Result'])
data1.head()
```

```python
# 데이터 분리
# 문제 데이터(X)와 정답 데이터(y) 분리
# 이해가 안되면 잊어도 된다
# 문제데이터를 연산하기 위해 백터로 변환
# 2차원 이상의 백터는 대문자 표기(X), 1차원 백터는 소문자 표기(y)

# X = 변하는 데이터
# y = 변하는 데이터로 인해 결정될 값
X = data1.loc[ : , 'N1':'N2']
y = data1.loc[ : , 'Result']
```

```python
# 머신러닝 모델 학습
# LinearRegression 모델 사용
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# 머신러닝 모델 학습
# x(문제 데이터)가 y(정답 데이터)를 도출하기 위한 규칙 찾기
# 정상적으로 실행되면 규칙이 찾아진 상태! -> 규칙이 맞는지 틀린지는 모름!
lr.fit(X,y)
```

```python
# 예측
lr.predict([[22,7]])
```

```python
# 머신러닝 모델이 찾은 규칙 살펴보기
# 모든 데이터에서 최대한 오류가 적은 규칙 한 가지만 찾아냄
lr.intercept_
```

```python
lr.coef_
```

```python
X
```

```python
# 다른 모델 사용하기
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
# 모델 학습하기
rf.fit(X,y)
# 예측
rf.predict([[22,7]])
```

```python
# 2개의 모델중 RandomForestRegressor 모델이 찾은 규칙이 더 잘 맞는거 같다
```

```python
# 규칙이 완벽하진 않아도 얼추 맞는다
rf.predict([[200,150]]) # 원래 정답 = 250, 출력 정답 = 100
# 학습한 데이터의 범위가 1~100 까지라서 그 이상의 값은 잘 모른다
```

# Q2.
- 4 ? 19 = 3
- (큰수/작은수)의 나머지

```python
 # 데이터 수집하기(만들기)
list2 = []
for i in range(1,101):
    for j in range(1,101):
        result = (1%j) if i>j else (j%i)
        list2.append([i,j,result])
```

```python
# 데이터 전처리
# 머신러닝 모델에 학습 가능한 형태로 데이터 변경, 목적에 따라 수행할게 달라진다.
# 연산이 가능한 형태로 타입 변경하기
import pandas as pd
data2 = pd.DataFrame(list2, columns = ['N1', 'N2', 'Result'])
```

```python
# 데이터 분리
# 문제와 정답 분리하기
X = data2.loc[:,'N1':'N2']
y = data2.loc[:,'Result']
```

```python
X
```

```python
# 머신러닝 모델 불러오기
# 사용할 데이터와 목적에 맞는 모델 선택하기
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
# 모델 학습하기
# 문제데이터로 정답데이터를 도출하기 위한 규칙 찾기
# 머신러닝 모델마다 접근하는 방법이 다르다.
rf.fit(X,y)
```

```python
# 예측하기
# 모델 학습시 마든 규칙으로 새로운 문제 데이터에 대한 정답 도출해보기
rf.predict([[4,19]])
```

```python
# 모델이 얼마나 잘 학습되었는지 수치적으로 확인하기
# 모델 자체적으로 판단한 학습의 완성도, 0~1, 1 :, 완벽 0 : 엉망
rf.score(X,y)
```

# 머신러닝
- 문제 데이터가 정답 데이터를 도출 하기 위한 규칙을 찾자
- 데이터에 따라 전처리가 달라진다
- 모델에 따라서 결과가 달라진다

# 머신러닝 코드의 흐름
- 1. 머신러닝으로 풀어낼 문제 찾기 = 미스터리 사인 규칙 찾기
- 2. 데이터 수집하기 -> 직접 만들어서 해결
-  3. 데이터 전처리 = 머신러닝 모델에 학습 가능한 형태로 변경. pd.DataFame
-  4. 데이터 분리 = 문제(X)와 정답(y)으로 데이터 나누기
-  5. 모델 불러오기 = RandomForest 모델 불러오기
-  6. 모델 학습하기 = 문제 데이터로 정답 데이터 도출하기 위한 규칙 찾기
-  7. 예측 및 평가 = 찾아진 규칙으로 새로운 문제 데이터의 결과 예측하기, 모델 스스로 규칙 평가

```python
list2 = []
for i in range(1,101):
    for j in range(1,101):
        result = (i%j) if i>j else (j%i)
        list2.append([i,j,result])
import pandas as pd
data2 = pd.DataFrame(list2, columns = ['N1','N2','Result'])
######################### 아래 코드는 거의 변하지 않음 #########################
X = data2.loc[: , 'N1':'N2']
y = data2.loc[:, 'Result']
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X,y)
print(rf.predict([[4,19]]))
print(rf.score(X,y))
```

# 지도학습
- 문제가 정답을 출력하기 위한 규칙 찾기
- 분류/ 회귀 두가지 종류가 존재(정답 데이터의 형태에 따라 달라짐)
   ### * 분류!
- 정답의 종류가 정해져 있는 문제
- 예) 혈액형, MBTI(정답의 종류가 정해져 있어서 찍는게 가능하다)
- 정답이 범주형 데이터로 이루어짐


  
   ### * 회귀
- 예측할 정답의 범위가 어느정도만 있는 문제
- 예) 집값, 강수량, 연봉
- 조그마한 차이가 크게 중요하지 않음
- 정답이 연속형 데이터로 이루어짐(수치 데이터)

  ### 지도학습
  - 분류, 회귀
  ### 비지도 학습
  - 군집, 차원축소

# 비지도 학습
- 정답데이터가 없이 학습을 하는 방법, 문제 데이터만 존재
- 문제 데이터 자체의 특징, 구조, 패턴을 찾는다
  
   
- * 군집 : 비슷한 특징이 있는 데이터들 끼리 그룹짓기

  -> 이미지 감색처리, 소비자 그룹 찾기, 비슷한 데이터끼리 묶어주기
- * 차원축소 : 데이터의 특징을 유지한채로 데이터 크기 줄이기
  
  -> 데이터의 핵심을 유지하려 한다

# 데이터를 문제, 정답으로 분리

```python
X=data.iloc[:,1:]
y=data['poisonous']
```

### 인코딩
- 레이블 인코딩(Label Encoding)
  - 특정 문자 데이터를 0,1,2 ...순으로 정수 간격으로 매핑하는 방식
  - 숫자의 크고 작음에 대한 특성이 작용하므로 거리나 수치 기반 모델에 적용했을 때 정확한 예측을 하기 힘들 수 있음
- 원 핫 인코딩(One-Hot Encoding)
  - 특정 컬럼의 유니크 값에 대한 더미 컬럼들을 만들어 실제 데이터가 있는 자리에 1로 표시하고 나머지는 0으로 채워주는 방식
  - 컴퓨터는 모든 데이터를 2진수로 처리하기 때문에 레이블 인코딩에 비해 기계가 직관적으로 이해하기 쉬우며, 수치의 크고 작음에 대한 특성이 없음
  - 데이터의 유니크 값들이 많은 경우 그 만큼 더미 컬럼들이 만들어져야 하기 때문에 메모리 소모가 클 수 있음

```python
X['cap-shape'].value_counts()
```

```python
X['cap-shape'].map({'x':0,'f':1,'k':2,'b':3,'s':4,'c':5})
```

```python
#레이블 인코딩을 자동으로 해주는 레이블인코더 임포트
from sklearn.preprocessing import LabelEncoder

# 한 번에 하나의 특성(1차원)만 변환이 가능(모든 데이터를 한번에 인코딩을 시도하면 특성들간 독립성이 깨어져서 예측력이 저하될 수 있음)
```

```python
encoder= LabelEncoder()  # 레비을 인코더 객체 생성
encoder.fit(X['cap-shape']) # 레이블 인코더로 학습(객체에게 우리의 범위를 알려주는 작업)
encoder.transform(X['cap-shape'])  #학습된 인코더 객체로 실제로 수치 값으로 변환
```

```python
# 인코더가 변환한 값들의 순서 확인(인덱스 번호로 변환됨)
encoder.classes_
```

### 원 핫 인코딩

```python
X['cap-shape']
```

```python
# get_dummies : pandas에서 지원해주는 원핫인코딩 함수
temp = pd.get_dummies(X[['cap-shape']],dtype = int)
temp
```

```python
X_one_hot =pd.get_dummies(X,dtype=int)
X_one_hot
```

- 정답 데이터의 경우 기계가 학습하고, 모델이 예측한 정답과 test 데이터의 정답을 비교만 하기 때문에 인코딩을 할 필요가 없음

```python
y_one_hot =pd.get_dummies(y,dtype=int)
y_one_hot           # 정답은 안해도됨 
```

## 5. 모델링

```python
# 학습 및 평가 데이터로 분리해보세요 (7:3) - 데이터는 원 핫 인코딩 데이터 기반으로 !!!
X_train, X_test,y_train,y_test = train_test_split(X_one_hot,y,
                                                 test_size=0.3,   
                                                 random_state=10)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### 결정트리 모델을 불러와서 학습, 예측, 평가 진행

```python
tree_model = DecisionTreeClassifier()
# 트리모델은 하이퍼파라미터 설정을 하지 않으면 완전히 분류될 떄 까지(리프노드의 불순도가 0)이 될 때까지 분기함
tree_model.fit(X_train,y_train)
```

```python
tree_model.score(X_train,y_train)
```

```python
tree_model.score(X_test,y_test)
```

### 학습된 트리 모델의 내부 구조를 시각화하기

```python
# max_depth 에 따른 train, test 데이터의 정확도 구해서 비교하기
train_list = []
test_list = []

for i in range(1,11,1) :
    tree_model3 = DecisionTreeClassifier(max_depth=i)
    # 모델 학습
    tree_model3.fit(X_train,y_train)

    #정확도 측정
    train_score = tree_model3.score(X_train,y_train)
    train_list.append(train_score)

    test_score = tree_model3.score(X_test,y_test)
    test_list.append(test_score)
```

```python
plt.figure(figsize=(15,5))
plt.plot(range(1,11), train_list, label='train')
plt.plot(range(1,11), test_list, label='test')
plt.xticks(range(1,11))
plt.legend();
```

### 특성 선택(feafure Selection)
- 정답을 예측함에 있어서 문제 데이터 특성들의 중요도를 수치로 출력할 수 있음(즉, 해당 특성이 정답을 예측하는데 얼마나 기여했는지를 알 수 있음)
  - 각 특성들은 0~1사이의 중요도 값을 가지며 모든 특성들의 중요도의 합은 1이 됨
- 0이 나오는 특성은 모델이 정답을 예측하는데 전혀 도움이 되지 않은 특성(없어도 무관)
- 1은 완벽하게 정답을 예측해 내는 특성
  - 1인 특성이 있다면 모델이 정답을 예측하는데 해당 특성만으로 유추를 했다는 뜻( 즉 다른 특성들은 정답예측에 필요가 없는 상태)

```python
fi = tree_model.feature_importances_
print(fi)
```

```python
len(fi)
```

```python
# 특성들의 중요도를 DF로 출력해보자
fi_df =pd.DataFrame(fi, index=X_one_hot.columns, columns=['특성중요도'])
fi_df
```

```python
fi_df.sort_values(by='특성중요도',ascending=False,inplace=True)
fi_df
```

- tree 모델 내부 시각화 표를 보면 가장 중요도가 높은 odor_n을 가장 먼저 고려해서 분류한 것을 알 수 있음
- 즉, tree 모델은 중요도가 높은 특성을 먼저 고려해야 효율적으로 최종 분류를 해낼 수 있음
- 현재 mushroom 데이터는 odor_n특성이 60%이상의 중요도를 혼자 차지하고 있기 때문에 기계학습이 매우 용이하여 100% 정확도를 얻어내기가 쉬웠음(애초에 분류하기 쉬운 데이터 였다는 뜻!)
